```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

# Simulation results 1

```python
df_1 = pd.read_csv('simulation_results_1.csv', header=1)
df_1.head(10)
```

```python
df_1.rename(columns={'Unnamed: 0': 'Date', 'Unnamed: 1': 'Time', 'Dry-bulb temperature (�C)': 'Dry-bulb temperature (°C)'}, inplace=True)
df_len_1 = len(df_1["Date"])
# Drop the first 5 rows
df_1 = df_1.drop(df_1.index[:5])

df_1.head(10)
```

```python
# Custom date parser function
def custom_date_parser(date_str):
    return pd.to_datetime(date_str, format='%a, %d/%b')

# Convert the Date column using the custom parser
df_1["Date"] = df_1["Date"].apply(custom_date_parser)
df_1["Date"] = df_1["Date"].dt.strftime('%m-%d')
```

```python
for i in range(0, len(df_1["Date"]), 144):
    # Get the date from row 0
    date_to_duplicate = df_1.iloc[i]['Date']

    # Assign this date to rows 0 to 143
    df_1.iloc[i:144 + i, df_1.columns.get_loc('Date')] = date_to_duplicate
```

```python
for i in range(6, len(df_1["Date"]), 6):
    temp_to_duplicate = df_1.iloc[i]['Dry-bulb temperature (°C)']
    
    df_1.iloc[i - 4, df_1.columns.get_loc('Dry-bulb temperature (°C)')] = temp_to_duplicate
    df_1.iloc[i - 2, df_1.columns.get_loc('Dry-bulb temperature (°C)')] = temp_to_duplicate
```

```python
for i in range(6, df_len_1, 2):
    df_1.drop(i, inplace=True)
```

```python
for i in range(5, df_len_1, 6):
    df_1.drop(i, inplace=True)
```

```python
df_1.iloc[-1, df_1.columns.get_loc('Dry-bulb temperature (°C)')] = df_1.iloc[-3]['Dry-bulb temperature (°C)']
df_1.iloc[-2, df_1.columns.get_loc('Dry-bulb temperature (°C)')] = df_1.iloc[-3]['Dry-bulb temperature (°C)']
```

```python
# Convert Date and Time columns to datetime objects
df_1['Date'] = pd.to_datetime(df_1['Date'], format='%m-%d')
df_1['Time'] = pd.to_datetime(df_1['Time'], format='%H:%M').dt.time

# Combine Date and Time columns into a single datetime column
df_1['DateTime'] = df_1.apply(lambda row: datetime.combine(row['Date'], row['Time']), axis=1)

# Format the combined datetime column to the desired format
df_1['DateTime'] = df_1['DateTime'].dt.strftime('%m-%d %H:%M')
```

```python
# Drop the Date and Time columns
df_1.drop(columns=['Date', 'Time'], inplace=True)
# Rearrange 'DateTime' column to be the first column
column_to_move = df_1.pop('DateTime')
df_1.insert(0, 'DateTime', column_to_move)
```

```python
df_1.reset_index(drop=True, inplace=True)
# Export the DataFrame to a CSV file
df_1.to_csv('simulation_results_1_completed.csv', index=False)
```
