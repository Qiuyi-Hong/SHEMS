```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from datetime import datetime, timedelta
```

# Monthly Cost Results

```python
cost_monthly = pd.read_csv('cost_results_monthly.csv')
cost_monthly.head()
```

```python
cost_monthly.set_index('From', inplace=True)
cost_monthly.head()
```

```python
fig, ax1 = plt.subplots()
lineStyle = ["-", "-.", "--", ":"]    
patterns = [ "/" , "\\", "*" , "+" , "-", "o", "O", "x", "*", "|"]

time_range = cost_monthly.index

ax1.plot(time_range, np.array(cost_monthly["Cost_HP"]), label="Cost HP", color="#1B4F72", linestyle=lineStyle[0], linewidth=3)

ax1.plot(time_range, np.array(cost_monthly["Cost_Gas_Boiler"]), label="Cost Gas Boiler", color="#B8860B", linestyle=lineStyle[0], linewidth=3)


plt.xticks(rotation=45)  # Rotate labels to avoid overlap

ax1.set_xlabel("Time (h)")
ax1.set_ylabel("Cost (p)")


# Collect handles and labels
handles1, labels1 = ax1.get_legend_handles_labels()

# Create a single legend
ax1.legend(handles1, labels1, loc='lower center', bbox_to_anchor=(0.5, 1.05), ncol=2)

plt.savefig(f"./figures/cost_results_monthly.pdf", bbox_inches='tight')
plt.show()
```
